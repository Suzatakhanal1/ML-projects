```python
data = data.sample(frac = 0.1, random_state = 1)
data.shape
```

```python
#plot histogram of each parameter
data.hist(figsize = (20,20))
plt.show()
```

```python
# determine no.of fraud cases in dataset
Fraud = data[data['Class']==1]
Valid = data[data['Class']==0]
outlier_fraction = len(Fraud)/ float(len(Valid))
print(outlier_fraction)
print(Fraud.shape) 
print(Valid.shape)
```

```python
#correlation metrics
corr = data.corr()
fig = plt.figure(figsize = (12,9))
sns.heatmap(corr)
plt.show()
```

```python
#get all the columns from df
columns = data.columns.tolist()

#filter the columns to remove data we do not want
columns = [c for c in columns if c not in ["Class"]]

#store the variable we'll be predicting on

target = "Class"

X = data[columns]
Y = data[target]
print(X.shape)
print(Y.shape)
```

```python
X = data[columns]
X
```

```python
#apply algorithms
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
```

```python
# define a model and put in dictyionary
classifiers = {
    
    "isolation_forest" : IsolationForest(contamination = outlier_fraction, random_state = 1),
    "local_outlier_factor" : LocalOutlierFactor(n_neighbors = 20,contamination = outlier_fraction)
}
```

```python
ifo = IsolationForest(contamination = outlier_fraction, random_state = 1)
ifo_pred = ifo.fit_predict(X)
#Reshape the prediction values to 0 for Valid transactions , 1 for Fraud transactions
ifo_pred[ifo_pred==1] = 0
ifo_pred[ifo_pred==-1] = 1
ifo_score = accuracy_score(Y,ifo_pred)
ifo_score
```

```python
lof = LocalOutlierFactor(n_neighbors = 20,contamination = outlier_fraction)
lof_pred = lof.fit_predict(X)
#Reshape the prediction values to 0 for Valid transactions , 1 for Fraud transactions
lof_pred[ifo_pred==1] = 0
lof_pred[ifo_pred==-1] = 1
lof_score = accuracy_score(Y,lof_pred)
lof_score
```

```python
print(classification_report(Y, lof_pred))
```
